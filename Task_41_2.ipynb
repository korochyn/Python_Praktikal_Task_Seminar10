```python
import seaborn as sns
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head(0)
```

```python
sns.scatterplot(data = df, x = 'longitude', y = 'latitude', hue = 'population')
```

```python
col = ['population', 'median_income', 'housing_median_age', 'median_house_value']
g = sns.PairGrid(df[col])
g.map(sns.scatterplot)
```

```python
sns.histplot(data = df[(df ['population'] > 0) & (df['population'] <= 5000)], x = 'population')
```
